```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:taniya05@127.0.0.1:3306/zomato_analytics"
)

users = pd.read_sql("SELECT * FROM users", engine)
orders = pd.read_sql("SELECT * FROM orders", engine)
restaurant = pd.read_sql("SELECT * FROM restaurant", engine)
food = pd.read_sql("SELECT * FROM food", engine)
menu = pd.read_sql("SELECT * FROM menu_clean", engine)
delivery = pd.read_sql("SELECT * FROM delivery_operations", engine)

print(restaurant.head())
```

```python
# Insight 1

top_city = restaurant["city"]\
.value_counts()\
.head(1)

print(top_city)

# Insight 2

highest_rated_city = restaurant.groupby(
    "city"
)["rating"]\
.mean()\
.sort_values(
    ascending=False
).head(10)

print(highest_rated_city)

# Insight 3

expensive_cuisine = menu.groupby(
    "cuisine"
)["price"]\
.mean()\
.sort_values(
    ascending=False
).head(10)

print(expensive_cuisine)

# Insight 4

weather_delay = delivery.groupby(
    "Weather_conditions"
)["Time_taken (min)"]\
.mean()\
.sort_values(
    ascending=False
)

print(weather_delay)

# Insight 5

traffic_delay = delivery.groupby(
    "Road_traffic_density"
)["Time_taken (min)"]\
.mean()\
.sort_values(
    ascending=False
)

print(traffic_delay)

# Insight 6

vehicle_performance = delivery.groupby(
    "Type_of_vehicle"
)["Time_taken (min)"]\
.mean()\
.sort_values()

print(vehicle_performance)

# Insight 7

festival_impact = (
    delivery.groupby("Festival")["Time_taken (min)"]
    .mean()
)

print(festival_impact)
```

```python
restaurant.to_csv("restaurant.csv", index=False)

orders.to_csv("orders.csv", index=False)

food.to_csv("food.csv", index=False)

menu.to_csv("menu.csv", index=False)

delivery.to_csv("delivery.csv", index=False)
```
